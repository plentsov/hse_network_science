### Task 1.

Load [wordcounts](http://www.leonidzhukov.net/hse/2015/networks/data/wordcounts.txt) dataset. 
1. Check that Zipf's Law holds
2. Assuming that the data is distributed according to the Power Law, find
 * $\alpha$ of the distribution
 * mean sample variance $\sigma^2$
3. Produce summary of the frequencies: min, max, mean, median

<b>1.</b> Let's see on the word frequencies distribution and see if it follows power law.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.xkcd()
import networkx as nx
%matplotlib inline
import numpy as np
```

```python
wordcounts = pd.read_csv('wordcounts.txt', sep=' ', header=None)
wordcounts.columns = ['count', 'word']
```

```python
wordcounts.head()
```

```python
plt.scatter(wordcounts.index, wordcounts['count'])
plt.xlabel("Word's rank")
plt.ylabel("Word's frequency")
plt.title("Here we see something like power law. Let's log!")
plt.show()
plt.show()
```

```python
wordcounts['log_count'] = np.log(wordcounts['count'])
wordcounts['log_rank'] = np.log(wordcounts.index)
```

```python
plt.scatter(wordcounts['log_rank'], wordcounts['log_count'])
plt.xlabel("log(Word's rank)")
plt.ylabel("log(Word's frequency)")
plt.title("Wordcounts dataset: Zipf's law")
plt.show()
```

Now it really looks like power law, because we have a linear relationship in log-coordinates<br/>
<b>2.</b> From the lectures we know, that maximum likelihood estimates for power law parameter and error are:<br/>
$$\gamma = 1 + n[\sum_{i=1}^{n}\frac{x_i}{x_{min}} ]^{-1}$$ (in our case $\gamma$ is referred to as $\alpha$)<br/>
$$\sigma = \frac{\gamma-1}{\sqrt{n}}$$ <br/>


```python
n = len(wordcounts)
x_min = min(wordcounts['count'])

alpha = 1 + n / sum(np.log(wordcounts['count']*1.0/x_min))
sigma_2 = ((alpha-1)/np.sqrt(n))**2

print "alpha =", round(alpha,2)
print "sigma^2 =", round(sigma_2,4)
```

<b>3.</b> Now let's look at summary statistics:

```python
print 'Min =', min(wordcounts['count'])
print 'Max =', max(wordcounts['count'])
print 'Mean =', round(np.mean(wordcounts['count']),2)
print 'Median =', np.median(wordcounts['count'])
```

# Routing network

```python
r_edges = np.loadtxt("network.txt", skiprows=4, dtype="i8,i8")
```

```python
r_G = nx.Graph()
r_G.add_edges_from(r_edges)
```

```python
r_degrees = r_G.degree().values()
r_n = len(r_degrees)
```

```python
r_degrees_dist = pd.value_counts(r_degrees)
r_degrees_dist.sort_index(inplace = True)
```

```python
plt.scatter(np.log(r_degrees_dist.index), np.log(r_degrees_dist))
plt.xlabel("log(degree)")
plt.ylabel("log(# vertices)")
plt.show()
```

```python
r_pdf = r_degrees_dist/r_n
plt.scatter(r_degrees_dist.index, r_pdf)
plt.title('Degree dist pdf')
plt.show()
```

```python
r_cdf = np.cumsum(r_pdf)
plt.scatter(r_degrees_dist.index, r_cdf)
plt.title('CDF')
plt.show()
```

We've seen this before, it's power law :)

In the network description it is said that graph is undirected. So we are going to demonstrate just minimum and maximum values of degrees here

```python
print "Mean degree =", round(np.mean(r_degrees),2)
print "Max degree =", max(r_degrees)
```

Now we're going to calculate maximum likelihood estimate for distribution parameters:

```python
r_x_min = min(r_degrees)
r_degrees = np.array(r_degrees)

r_alpha = 1 + r_n / sum(np.log(r_degrees*1.0/r_x_min))
r_sigma_2 = ((r_alpha-1)/np.sqrt(r_n))**2

print "alpha =", round(r_alpha,2)
print "sigma^2 =", round(r_sigma_2,4)
```

Next step is to find optimal $x_{min}$ using Kolmogorov-Smirnov test:

```python
r_range = np.arange(1, 300, 1)
r_x_min_opt = r_x_min
r_d_min = np.infty

for x in r_range:
    constant = (r_alpha - 1)*(x**(r_alpha-1))
    # as we suppose that power law starts with k = x, we will drop off vertices with lower degrees
    model_pdf = [(r_alpha-1)/x*((k/x)**(-r_alpha)) for k in r_degrees_dist.index if k >= x] 
    model_pdf = model_pdf/sum(model_pdf)
    model_cdf = np.cumsum(model_pdf)
    exp_pdf = r_degrees_dist[r_degrees_dist.index >= x]
    exp_pdf = exp_pdf*1.0/sum(exp_pdf)
    exp_cdf = np.cumsum(exp_pdf)
    d = max(abs(model_cdf - exp_cdf))
    if d < r_d_min: 
        r_d_min = d
        r_x_min_opt = x
```

```python
print r_x_min_opt
```

So, we say that a power law starts with a degree 4.

This is like a Poisson one, not a power law :(.

# Web graph

```python
w_edges = np.loadtxt("web_Stanford.txt", skiprows=4, dtype="i8,i8")
```

```python
w_G = nx.DiGraph()
w_G.add_edges_from(w_edges)
```

```python
w_in_degrees = np.array(w_G.in_degree().values())
w_in_degrees = w_in_degrees[w_in_degrees > 0]
w_out_degrees = np.array(w_G.out_degree().values())
w_out_degrees = w_out_degrees[w_out_degrees > 0]
w_in_n = len(w_in_degrees)
w_out_n = len(w_out_degrees)
```

```python
w_in_degrees_dist = pd.value_counts(w_in_degrees)
w_in_degrees_dist.sort_index(inplace = True)

w_out_degrees_dist = pd.value_counts(w_out_degrees)
w_out_degrees_dist.sort_index(inplace = True)
```

```python
w_in_pdf = w_in_degrees_dist/w_in_n
plt.scatter(w_in_degrees_dist.index, w_in_pdf)
plt.title('In-Degree dist pdf')
plt.show()
```

```python
w_out_pdf = w_out_degrees_dist/w_out_n
plt.scatter(w_out_degrees_dist.index, w_out_pdf)
plt.title('In-Degree dist pdf')
plt.show()
```

Both distributions look like a power law!

```python
print "Mean in-degree =", round(np.mean(w_in_degrees),2)
print "Max in-degree =", max(w_in_degrees)
print
print "Mean out-degree =", round(np.mean(w_out_degrees),2)
print "Max out-degree =", max(w_out_degrees)
```

While mean degrees are almost the same, maximum out degree is much smaller than maximum in-degree. So some vertice has a lot of in-edges.

Now let's calculate ML estimators for distributions' parameters

```python
w_in_x_min = min(w_in_degrees)
w_in_degrees = np.array(w_in_degrees)

w_in_alpha = 1 + w_in_n / sum(np.log(w_in_degrees*1.0/w_in_x_min))
w_in_sigma_2 = ((w_in_alpha-1)/np.sqrt(w_in_n))**2

print "In-degree dist. alpha =", round(w_in_alpha,2)
print "In-degree dist. sigma^2 =", round(w_in_sigma_2,10)

print
w_out_x_min = min(w_out_degrees)
w_out_degrees = np.array(w_out_degrees)

w_out_alpha = 1 + w_out_n / sum(np.log(w_out_degrees*1.0/w_out_x_min))
w_out_sigma_2 = ((w_out_alpha-1)/np.sqrt(w_out_n))**2

print "Out-degree dist. alpha =", round(w_out_alpha,2)
print "Out-degree dist. sigma^2 =", round(w_out_sigma_2,10)
```

Next step is to find optimal $x_{min}$ using Kolmogorov-Smirnov test:

```python
# In-degree dist estimation:

w_in_range = np.arange(1, 300, 1)
w_in_x_min_opt = w_in_x_min
w_in_d_min = np.infty

for x in w_in_range:
    constant = (w_in_alpha - 1)*(x**(w_in_alpha-1))
    # as we suppose that power law starts with k = x, we will drop off vertices with lower degrees
    model_pdf = [(w_in_alpha-1)/x*((k/x)**(-w_in_alpha)) for k in w_in_degrees_dist.index if k >= x]/sum(model_pdf)
    model_cdf = np.cumsum(model_pdf)
    exp_pdf = w_in_degrees_dist[w_in_degrees_dist.index >= x]
    exp_pdf = exp_pdf*1.0/sum(exp_pdf)
    exp_cdf = np.cumsum(exp_pdf)
    d = max(abs(model_cdf - exp_cdf))
    if d < w_in_d_min: 
        w_in_d_min = d
        w_in_x_min_opt = x
        
print "Optimum value of x_min for in-degree dist:", w_in_x_min_opt


# Out-degree dist estimation:

w_out_range = np.arange(1, 200, 1)
w_out_x_min_opt = w_out_x_min
w_out_d_min = np.infty

for x in w_out_range:
    constant = (w_out_alpha - 1)*(x**(w_out_alpha-1))
    # as we suppose that power law starts with k = x, we will drop off vertices with lower degrees
    model_pdf = [(w_out_alpha-1)/x*((k/x)**(-w_out_alpha)) for k in w_out_degrees_dist.index if k >= x]/sum(model_pdf)
    model_cdf = np.cumsum(model_pdf)
    exp_pdf = w_out_degrees_dist[w_out_degrees_dist.index >= x]
    exp_pdf = exp_pdf*1.0/sum(exp_pdf)
    exp_cdf = np.cumsum(exp_pdf)
    d = max(abs(model_cdf - exp_cdf))
    if d < w_out_d_min: 
        w_out_d_min = d
        w_out_x_min_opt = x
        
print "Optimum value of x_min for out-degree dist:", w_out_x_min_opt
```
