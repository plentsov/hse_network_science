### Task 1

Consider Barabasi and Albert dynamical grow model. Two main ingredients of this model are *network growing* and *prefferential attachment*. Implement two restricted B&A-based models:
<br \>

**Model A**
<br \>
Lack of prefferential attachment, that is at each time-step form edges uniformly at random while network keeps growing.

Such restricted models are described <a href = "http://barabasi.com/f/622.pdf">here</a> (pp.16-17) <br/>

```python
# n_0 - initial number of nodes in the network
# m - number of edges to be added with new node
# T - number of timesteps

def BA_growth_only(n_0, m, T):
    number_of_nodes = []
    path_lens = []
    cluster_coefs = []

    # creating initial network
    p = 1.0/n_0
    net = nx.fast_gnp_random_graph(n_0, p, seed = 18)
    while ((net.number_of_edges() < n_0) or (not nx.is_connected(net))):
        p *= 3
        net = nx.fast_gnp_random_graph(n_0, p, seed = 18)
    number_of_nodes.append(net.number_of_nodes())
    path_lens.append(nx.average_shortest_path_length(net))
    cluster_coefs.append(nx.average_clustering(net))

    # growth
    for t in range(T):
        net.add_node(n_0 + t)
        nodes_to_connect = np.random.choice(range(n_0 + t), m, replace=False)
        for v in nodes_to_connect:
            net.add_edge(n_0 + t, v)

        if (n_0+t)%50 == 0:
            number_of_nodes.append(net.number_of_nodes())
            path_lens.append(nx.average_shortest_path_length(net))
            cluster_coefs.append(nx.average_clustering(net))
        
    return {'n_0': n_0, 'm': m, 'T' : T, 'net': net, 
            'nn': number_of_nodes, 'path_lens': path_lens, 'ccs': cluster_coefs}
```

Let's implement several instances (with different inputs) of this model and compare their results.

```python
s = datetime.now()
models_A = []
models_A.append(BA_growth_only(3, 2, 1097))
print "First model built, time elapsed:", datetime.now() - s
s = datetime.now()
models_A.append(BA_growth_only(100, 20, 1000))
print "Second model built, time elapsed:", datetime.now() - s
s = datetime.now()
models_A.append(BA_growth_only(100, 50, 1000))
print "Third model built, time elapsed:", datetime.now() - s
```

```python
l = []
colors = ['blue', 'red', 'green']
plt.title('Degree PDFs of growth-only models')

for i in range(3):
    sns.distplot(models_A[i]['net'].degree().values(), color=colors[i])
    l.append('$n_0 ='+str(models_A[i]['n_0'])+', m = '+str(models_A[i]['m'])+'$')

plt.legend(l)
plt.ylim([0, 0.08])
plt.show()
```

This distribution does not seem like a power law, it is definitely not scale-free, there are no hubs in the network.<br/>
It is more like an exponential one.

```python
plt.title('Degree CDFs of growth-only models')

for i in range(3):
    sns.kdeplot(np.array(models_A[i]['net'].degree().values()), cumulative=True, legend=False, color=colors[i])
    
plt.legend(l)
plt.show()
```

```python
plt.title('Avg path length of growth-only models')
for i in range(3):
    plt.plot(models_A[i]['nn'], models_A[i]['path_lens'], color=colors[i])
plt.xlabel('# Nodes')
plt.legend(l)
plt.show()
```

```python
plt.title('Avg clustering coefficient of growth-only models')
for i in range(3):
    plt.plot(models_A[i]['nn'], models_A[i]['ccs'], color = colors[i])
plt.xlabel('# Nodes')
plt.legend(l)
plt.show()
```

Clustering coefficient is low with low $m$, while with large $m$ we have seen that the degree distribution is far away from power law.

```python
for i in range(3):
    T = models_A[i]['T']
    n_0 = models_A[i]['n_0']
    ind = range(T)
    ind.extend([T]*n_0)
    degrees = models_A[i]['net'].degree().values()
    degrees.reverse()
    plt.scatter(ind, degrees, edgecolors='black', facecolors=colors[i])
    
plt.xlim([0, T+n_0])
plt.ylim([0, 200])
plt.xlabel("Node 'age'")
plt.ylabel('Degree')
plt.title('Node degree depending on age in growth-only models')
plt.legend(l, loc=2)
plt.show()
```

We see that older nodes tend to have higher degree

**Model B**
<br \>
Lack of growing, that is fix total number of nodes, on each time-step randomly choose one and form edges with prefferential attachment.
<br \>

1. Generate networks according to the models above ($N > 1000$ nodes)
2. Compute CDF/PDF, describe the distribution and compute\describe its properties.
3. Illustate the following dependencies: 
    * average path length to the number of nodes
    * average clustering coefficient to the number of nodes
    * average node degee to the nodes "age"
4. Is scale-free property conserved in these models?

Analyse results with respect to various parameter settings

```python
# n - fixed total number of nodes in the network
# m - according to the article at each timestamp we connect only one node
# T - number of timesteps
def BA_pref_att_only(n, T):
    m = 1
    number_of_edges = []
    path_lens = []
    cluster_coefs = []

    # initial net
    net = nx.empty_graph(n)
    number_of_edges.append(net.number_of_edges())
    path_lens.append(0)
    cluster_coefs.append(nx.average_clustering(net))
    ran = range(n)
    
    # prefferential attachment
    for t in range(T):
        random_node = np.random.randint(0,n)
        mod_degrees = np.array([max(1,d) for d in net.degree().values()])
        probas = mod_degrees*1.0/sum(mod_degrees)
        node_to_connect = np.random.choice(ran, size = m, replace = False, p = probas)
        net.add_edge(random_node, node_to_connect[0])

        if t%(T/20) == 0:
            path_lens.append(nx.average_shortest_path_length(max(nx.connected_component_subgraphs(net), key=len)))
            cluster_coefs.append(nx.average_clustering(net)) 
            number_of_edges.append(net.number_of_edges())
        
    return {'n_0': n_0, 'm': m, 'T' : T, 'net': net, 
            'ne': number_of_edges, 'path_lens': path_lens, 'ccs': cluster_coefs}
```

```python
s = datetime.now()
models_B = []
models_B.append(BA_pref_att_only(1000, 10**5))
print "First model built, time elapsed:", datetime.now() - s
#s = datetime.now()
#models_B.append(BA_pref_att_only(1000, 10**6))
#print "Second model built, time elapsed:", datetime.now() - s
```

```python
l = []
colors = ['blue', 'red']
plt.title('Degree PDFs of PrefAttach-only models')

for i in range(1):
    sns.distplot(models_B[i]['net'].degree().values(), color=colors[i])
    l.append('$n_0 ='+str(models_B[i]['n_0'])+', T = '+str(models_B[i]['T'])+'$')

plt.legend(l)
plt.show()
```

Obviously, not a power law case. Degrees are very limited due to the fact that all nodes have approximately the same probability to be chosen (as there are no "old" nodes)

```python
plt.title('Degree CDFs of PrefAttach-only models')

for i in range(1):
    sns.kdeplot(np.array(models_B[i]['net'].degree().values()), cumulative=True, legend=False, color=colors[i])
    
plt.legend(l)
plt.show()
```

```python
plt.title('Avg clustering coefficient of growth-only models')
for i in range(1):
    plt.plot(models_B[i]['ne'], models_B[i]['ccs'], color = colors[i])
plt.xlabel('# Edges added')
plt.legend(l)
plt.show()
```

Obviously, as we do not change number of nodes and only add new edge, clustering coefficient is growing. Seems like this growth is linear

So, again we do not have neither power law or scale-free property

### <hr /> Task 2

Consider the following  "Vertex copying model" of growing network.

At every time step a random vertex from already existing vertices is selected and duplicated together with all edges, such that  every edge of the  vertex
* is copied with probability $q$
* is rewired to any other randomly selected vertex with probability $1-q$


Starting state is defined by some small number of randomly connected vertices.

The model can generate both directed and undirected networks.

1. Generate  graphs based on the model ($N > 1000$ nodes)
2. Compute CDF/PDF, describe the distribution and compute\describe its properties.
3. Illustate the following dependencies: 
    * average path length to the number of nodes
    * average clustering coefficient to the number of nodes
    * average node degee to the nodes "age"
    
Analyse results with respect to various parameter settings

```python
# n_0 - initial number of nodes in the network
# T - number of timesteps
# q - probability that edge will be copied

def vertex_copying_model(n_0, T, q):
    number_of_nodes = []
    path_lens = []
    cluster_coefs = []

    # creating initial network
    p = 1.0/n_0
    net = nx.fast_gnp_random_graph(n_0, p, seed = 18)
    while ((net.number_of_edges() < n_0) or (not nx.is_connected(net))):
        p *= 3
        net = nx.fast_gnp_random_graph(n_0, p, seed = 18)
    number_of_nodes.append(net.number_of_nodes())
    path_lens.append(nx.average_shortest_path_length(net))
    cluster_coefs.append(nx.average_clustering(net))

    # growth
    for t in range(T):
        # which node to copy
        random_node = np.random.choice(range(n_0 + t), 1)[0]
        net.add_node(n_0 + t)
        nodes_to_connect = []
        random_node_neighbors = net.neighbors(random_node)
        for i in range(len(random_node_neighbors)):
            if np.random.rand() < q:
                nodes_to_connect.append(random_node_neighbors[i])
            else:
                nodes_to_rewire = set(range(n_0 + t)) - set(random_node_neighbors)
                node_to_rewire = np.random.choice(list(nodes_to_rewire), 1)[0]
                nodes_to_connect.append(node_to_rewire)
            
        for v in nodes_to_connect:
            net.add_edge(n_0 + t, v)

        if (n_0+t)%50 == 0:
            number_of_nodes.append(net.number_of_nodes())
            path_lens.append(nx.average_shortest_path_length(net))
            cluster_coefs.append(nx.average_clustering(net))
        
    return {'n_0': n_0, 'q': q, 'T' : T, 'net': net, 
            'nn': number_of_nodes, 'path_lens': path_lens, 'ccs': cluster_coefs}
```

```python
s = datetime.now()
models_2 = []
models_2.append(vertex_copying_model(100, 1000, 0))
print "First model built, time elapsed:", datetime.now() - s
s = datetime.now()
models_2.append(vertex_copying_model(100, 1000, 0.3))
print "Second model built, time elapsed:", datetime.now() - s
s = datetime.now()
models_2.append(vertex_copying_model(100, 1000, 0.7))
print "Third model built, time elapsed:", datetime.now() - s
s = datetime.now()
models_2.append(vertex_copying_model(100, 1000, 1))
print "Fourth model built, time elapsed:", datetime.now() - s
```

```python
l = []
colors = ['blue', 'red', 'green', 'yellow']
plt.title('Degree PDFs of vertex copying models')

for i in range(4):
    sns.distplot(models_2[i]['net'].degree().values(), color=colors[i])
    l.append('$n_0 ='+str(models_2[i]['n_0'])+', q = '+str(models_2[i]['q'])+'$')

plt.legend(l)
plt.ylim([0, 0.08])
plt.show()
```

Obviously, not power law, no scale-free property. It is clear that as $q$ grows, "peak" of the distribution vanishes, as more and more links are rewired, so it becomes more like uniform

```python
plt.title('Degree CDFs of vertex copying models')

for i in range(4):
    sns.kdeplot(np.array(models_2[i]['net'].degree().values()), cumulative=True, legend=False, color=colors[i])
    
plt.legend(l)
plt.show()
```

```python
plt.title('Avg path length of vertex copying models')
for i in range(4):
    plt.plot(models_2[i]['nn'], models_2[i]['path_lens'], color=colors[i])
plt.xlabel('# Nodes')
plt.legend(l)
plt.show()
```

If $q < 1$, average path length decreases with the growth of number of nodes, obviously. <br/>
If $q = 1$, it does not change, as new nodes essentialy duplicate existing ones, so the their shortest paths copy shortest paths of the existing nodes

```python
plt.title('Avg clustering coefficient of vertex copying models')
for i in range(4):
    plt.plot(models_2[i]['nn'], models_2[i]['ccs'], color = colors[i])
plt.xlabel('# Nodes')
plt.legend(l)
plt.show()
```

```python
for i in range(4):
    T = models_2[i]['T']
    n_0 = models_2[i]['n_0']
    ind = range(T)
    ind.extend([T]*n_0)
    degrees = models_2[i]['net'].degree().values()
    degrees.reverse()
    plt.scatter(ind, degrees, edgecolors='black', facecolors=colors[i])
    
plt.xlim([0, T+n_0])
plt.ylim([0, 200])
plt.xlabel("Node 'age'")
plt.ylabel('Degree')
plt.title('Node degree depending on age in vertex copying models')
plt.legend(l, loc=2)
plt.show()
```

As one could expect, node degree does not depend on its age in this model, as we add new node with a degree of some existing node. Moreover, we change older nodes degrees randomly without some preferrence. <br/>
As can be seen from this scatterplot, higher $q$ makes node degree variance higher, that was clear from PDF plot earlier.
