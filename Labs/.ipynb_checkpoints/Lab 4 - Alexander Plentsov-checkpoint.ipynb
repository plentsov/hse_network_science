During this seminar we will work with **Watts and Strogatz model**.
Again, the idea of the model:

1. Start with regular lattice with N nodes, K nearest neighbours
2. For each node take every edge and rewire it with probability $p$, assuming that there is no loops and edge duplications

```python
def gen_regular_graph(n, k):
    G = nx.Graph()
    nodes = list(range(n))
    for j in range(1, k // 2+1):
        targets = nodes[j:] + nodes[0:j] # first j nodes are now last in list
        G.add_edges_from(zip(nodes,targets))
    return G

def degree_distr(k, bins = 100):
    yh, binEdges = np.histogram(k, bins)
    bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
    return yh, bincenters
```

## 1. Function that rewires some nodes in regular graph 

```python
def rewind(G, p):
    n = len(G)
    for i in range(n):
        edges = G.edges(i)
        for e in edges:
            if np.random.rand() < p:
                G.remove_edge(*e)
                candidates = list(set(range(n)) - set(G.neighbors(i)) - set([i]))
                G.add_edge(i, np.random.choice(candidates))
    return G
```

## 2. Network properties with various p values

### p = 0

```python
G = gen_regular_graph(100, 4)
G = rewind(G, 0)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, node_size = 150, font_size=5)
```

```python
print "Average path length =", nx.average_shortest_path_length(G)
print "Clustering coefficient =", nx.transitivity(G)
```

```python
plt.hist(G.degree().values())
plt.title("Degree distribution with $p = 0$")
plt.ylabel("# nodes")
plt.xlabel("$k$")
plt.show()
```

### p = 1

```python
G = gen_regular_graph(100, 4)
G = rewind(G, 1)
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, node_size = 150, font_size=5)
```

```python
print "Average path length =", nx.average_shortest_path_length(max(nx.connected_component_subgraphs(G), key=len))
print "Clustering coefficient =", nx.transitivity(G)
```

```python
plt.hist(G.degree().values())
plt.title("Degree distribution with $p = 1$")
plt.ylabel("# nodes")
plt.xlabel("$k$")
plt.show()
```

So, here we got totally random graph

### Let's explore various $p$ values

```python
lens = []
trs = []
ps = []
for p in np.arange(0, 1.01, 0.05):
    G = gen_regular_graph(100, 4)
    G = rewind(G, p)
    ps.append(p)
    lens.append(nx.average_shbortest_path_length(max(nx.connected_component_subgraphs(G), key=len)))
    trs.append(nx.transitivity(G))
```

```python
fig, ax1 = plt.subplots()
ax1.plot(ps, lens, 'b')

ax2 = ax1.twinx()
ax2.plot(ps, trs, 'r')

ax1.set_xlabel('$p$')
ax1.set_ylabel('Avg path len', color='b')
ax2.set_ylabel('Clustering coef', color='r')
plt.show()
```

Seems like with $p$ around $0.1$ we have high enough clustering coefficient with moderate average path length.

## 4. Modelling political blogs net

```python
n_p = 1490
```

### Small world model

As a degree for each vertex we will use the average node degree in the real network:

```python
avg_deg = int(np.mean(G_p.degree().values()))
print avg_deg
```

Let's try to obtain the same clustering coefficient as in the real network. As we've seen earlier, clustering coefficient is going to decrease as $p$ grows.

```python
cc_real = nx.transitivity(G_p)
print cc_real
```

```python
cc = 1
p = 0
while cc > cc_real: 
    G_sw = gen_regular_graph(n_p, avg_deg)
    G_sw = rewind(G_sw, p)
    cc = nx.transitivity(G_sw)
    p += 0.02
```

```python
p -= 0.02
print p
```

### Erdos - Renyi Model

In order to fit Erdos-Renyi Model we need another $p$ - the probability that the edge exists.

```python
p_erdos = np.mean(G_p.degree().values())/n_p
print p_erdos
```

```python
G_er = nx.fast_gnp_random_graph(n_p, p_erdos)
```

## 5. Comparing centralities

```python
er_c_degree = nx.degree_centrality(G_er)
er_c_close = nx.closeness_centrality(G_er)
er_c_eigen = nx.eigenvector_centrality(G_er)
er_c_katz = nx.katz_centrality_numpy(G_er)
er_c_load = nx.load_centrality(G_er)

sw_c_degree = nx.degree_centrality(G_sw)
sw_c_close = nx.closeness_centrality(G_sw)
sw_c_eigen = nx.eigenvector_centrality(G_sw)
sw_c_katz = nx.katz_centrality_numpy(G_sw)
sw_c_load = nx.load_centrality(G_sw)
```
